#### read in data

```python
# pulls the stats for every player in the nba
tables = pd.read_html('https://www.basketball-reference.com/leagues/NBA_2025_per_game.html#per_game_stats')
df = tables[0]

#show
df
```

```python
# making a dictionary of the teams win proportions
dict_win = {'OKC': [0.829],
            'MIL': [0.585],
            'DEN': [0.610],
            'DAL': [0.476],
            'LAL': [0.610],
            'MIN': [0.598],
            'BOS': [0.744],
            'PHO': [0.439],
            'PHI': [0.293],
            'DET': [0.537],
            'NYK': [0.622],
            'ORL': [0.500],
            'CHO': [0.232],
            'NOP': [0.256],
            'GSW': [0.585],
            'SAS': [0.415],
            'ATL': [0.488],
            'CLE': [0.780],
            'BRK': [0.317],
            'MIA': [0.451],
            'SAC': [0.488],
            'CHI': [0.476],
            'MEM': [0.585],
            'LAC': [0.610],
            'TOR': [0.366],
            'HOU': [0.634],
            'WAS': [0.220],
            'IND': [0.610],
            'POR': [0.439],
            'UTA': [0.207]
}
df_win = pd.DataFrame(dict_win) # turning it into a data frame
df_win = df_win.melt(var_name='Team') # making it long format for the merge

# show
df_win
```

```python
# merge them together
df_final = pd.merge(left=df, right=df_win, how='left', left_on='Team', right_on='Team')
dict_rename = {'value': 'Team_Win'} # renames the 'value' column to 'Team_Win'
df_final = df_final.rename(dict_rename, axis=1)

# show
df_final
```

```python
# drop the rows that don't have a team win
df_final = df_final.dropna(subset=['Team_Win'])
df_final.reset_index(inplace=True) # resets the index after dropping

# show
df_final
```

```python
# save it to a csv
df_final.to_csv('./player_data.csv')
```
